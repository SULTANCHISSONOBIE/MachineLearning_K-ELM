```python
# Import Library
import numpy as np # linear algebra
import pandas as pd # data processing, pembacaan file excel
import matplotlib.pyplot as plt # untuk plotting
from sklearn import preprocessing # preprocessing untuk normalisasi
```

```python
# Pembacaan File Data Polusi Udara ISPU Excel
DKI1_FOR = pd.read_excel("Validasi_Forecasting_2.xlsx", sheet_name="DKI1_FOR")
DKI2_FOR = pd.read_excel("Validasi_Forecasting_2.xlsx", sheet_name="DKI2_FOR")
DKI3_FOR = pd.read_excel("Validasi_Forecasting_2.xlsx", sheet_name="DKI3_FOR")
DKI4_FOR = pd.read_excel("Validasi_Forecasting_2.xlsx", sheet_name="DKI4_FOR")
DKI5_FOR = pd.read_excel("Validasi_Forecasting_2.xlsx", sheet_name="DKI5_FOR")
```

```python
# Menampilkan 6 Data pertama DKI1
DKI1_FOR.head(6)
```

```python
# Hanya menampilkan data zat yang diukur
DKI1_FOR = DKI1_FOR.loc[:, ~DKI1_FOR.columns.isin(['Tanggal', 'Wilayah'])]
DKI2_FOR = DKI2_FOR.loc[:, ~DKI2_FOR.columns.isin(['Tanggal', 'Wilayah'])]
DKI3_FOR = DKI3_FOR.loc[:, ~DKI3_FOR.columns.isin(['Tanggal', 'Wilayah'])]
DKI4_FOR = DKI4_FOR.loc[:, ~DKI4_FOR.columns.isin(['Tanggal', 'Wilayah'])]
DKI5_FOR = DKI5_FOR.loc[:, ~DKI5_FOR.columns.isin(['Tanggal', 'Wilayah'])]
DKI1_FOR.head(20)
```

```python
# Normalisasi Data DKI1 sampai DKI5
scaler = preprocessing.MinMaxScaler()
names1 = DKI1_FOR.columns
names2 = DKI2_FOR.columns
names3 = DKI3_FOR.columns
names4 = DKI4_FOR.columns
names5 = DKI5_FOR.columns
d1 = scaler.fit_transform(DKI1_FOR)
d2 = scaler.fit_transform(DKI2_FOR)
d3 = scaler.fit_transform(DKI3_FOR)
d4 = scaler.fit_transform(DKI4_FOR)
d5 = scaler.fit_transform(DKI5_FOR)
DKI1_Normalization = pd.DataFrame(d1, columns=names1)
DKI2_Normalization = pd.DataFrame(d2, columns=names2)
DKI3_Normalization = pd.DataFrame(d3, columns=names3)
DKI4_Normalization = pd.DataFrame(d4, columns=names4)
DKI5_Normalization = pd.DataFrame(d5, columns=names5)
DKI1_Normalization.head(20)
```

```python
# Penyimpanan data yang sudah normalisasi ke dalam file csv
DKI1_Normalization.to_csv('C:/OBIE/Tel - U/SEM 7/PTA/BISMILLAH/Code/TA_COBA_KELM1/Forecasting/DKIFORECAST/DKI1_Normalization.csv')
DKI2_Normalization.to_csv('C:/OBIE/Tel - U/SEM 7/PTA/BISMILLAH/Code/TA_COBA_KELM1/Forecasting/DKIFORECAST/DKI2_Normalization.csv')
DKI3_Normalization.to_csv('C:/OBIE/Tel - U/SEM 7/PTA/BISMILLAH/Code/TA_COBA_KELM1/Forecasting/DKIFORECAST/DKI3_Normalization.csv')
DKI4_Normalization.to_csv('C:/OBIE/Tel - U/SEM 7/PTA/BISMILLAH/Code/TA_COBA_KELM1/Forecasting/DKIFORECAST/DKI4_Normalization.csv')
DKI5_Normalization.to_csv('C:/OBIE/Tel - U/SEM 7/PTA/BISMILLAH/Code/TA_COBA_KELM1/Forecasting/DKIFORECAST/DKI5_Normalization.csv')
```
