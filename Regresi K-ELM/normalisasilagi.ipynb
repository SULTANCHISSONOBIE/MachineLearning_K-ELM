```python
import pandas as pd
import statistics

# Fungsi Min-Max Scaling
def min_max_scaling(data):
    min_val = min(data)
    max_val = max(data)
    normalized_data = [(x - min_val) / (max_val - min_val) for x in data]
    return normalized_data

# Pembacaan File Data Polusi Udara ISPU Excel
DKI1 = pd.read_excel("Validasi_Forecasting_2.xlsx", sheet_name="DKI1")
DKI2 = pd.read_excel("Validasi_Forecasting_2.xlsx", sheet_name="DKI2")
DKI3 = pd.read_excel("Validasi_Forecasting_2.xlsx", sheet_name="DKI3")
DKI4 = pd.read_excel("Validasi_Forecasting_2.xlsx", sheet_name="DKI4")
DKI5 = pd.read_excel("Validasi_Forecasting_2.xlsx", sheet_name="DKI5")

# Normalisasi kolom tertentu di setiap DataFrame
columns_to_normalize = ["PM10", "SO2", "CO", "O3","NO2"]  # Ganti dengan nama kolom yang ingin Anda normalisasi

# Contoh normalisasi menggunakan Min-Max Scaling pada DKI1_FOR
for column in columns_to_normalize:
    DKI1[column] = min_max_scaling(DKI1[column])

# Cetak hasil DKI1_FOR setelah normalisasi
print("Hasil normalisasi DKI1:")
print(DKI1)
```
